```
jan = spark.read.csv('/FileStore/tables/Jan17.csv', header=True, inferSchema = True)
feb = spark.read.csv('/FileStore/tables/Feb17.csv', header=True, inferSchema = True)
display(feb)
```

```
#jan.groupBy("PULocationID").count().show()
#jan.select("DOLocationID").distinct().show()
vertexP = jan.select("PULocationID").distinct()
vertexP = vertexP.selectExpr("PULocationID as id")
print(vertexP.count())
vertexD = jan.select("DOLocationID").distinct()
vertexD = vertexD.selectExpr("DOLocationID as id")
print(vertexD.count())
if vertexD.count() > vertexP.count():
  vertex = vertexD.unionAll(vertexP)
else : vertex = vertexP.unionAll(vertexD)
print(vertex.count())
vertex = vertex.select("id").distinct()
print(vertex.count())
vertex.show()
```

```
Edge = jan.select('PULocationID', 'DOLocationID')
Edge.groupBy('PULocationID', 'DOLocationID')
Edge = Edge.selectExpr("PULocationID as src", "DOLocationID as dst")
edge = jan.groupBy('PULocationID', 'DOLocationID').count()
edge = edge.selectExpr("PULocationID as src", "DOLocationID as dst", "count as weight")
Edge.show()
```

```
from pyspark.sql import SQLContext
from pyspark import SparkContext
from graphframes import * 

graph = GraphFrame(vertex,Edge)
vertex.cache()
Edge.cache()
```

```
print("Total Number of Locations: " , graph.vertices.count())
print("Total Number of Trips in Graph: " , graph.edges.count())
print("Total Number of Trips in Original Data: " , Edge.count())
```

```
topTrips = graph.edges.groupBy("src", "dst").count()
topTrips.groupBy("count")
display(topTrips)
```

```
loc = spark.read.csv('/FileStore/tables/taxi__zone_lookup-3dc46.csv', header=True, inferSchema = True)
display(loc)
```

```
df_joined = topTrips.join(loc,[topTrips.src==loc.LocationID],'left_outer')
df_joined = df_joined.selectExpr("LocationID as PULocID", "Borough as PUBorough", "Zone as PUZone", "src as src", "dst as dst", "count as count")
df_joined = df_joined.join(loc,[df_joined.dst==loc.LocationID],'left_outer')
df_joined = df_joined.selectExpr("src as src", "dst as dst", "count as count", "PULocID as PULocID", "PUBorough as PUBorough", "PUZone as PUZone", "LocationID as DOLocID", "Borough as DOBorough", "Zone as DOZone")
df_joined.show()
```
